```python
import matplotlib.pyplot as plt

# Berilgan ma'lumotlar
spirt_miqdor = [365, 363, 366, 360, 361, 362, 358, 368, 359, 350, 346, 347, 346, 350, 354, 336]
havo_miqdor = [1351, 1344, 1340, 1341, 1345, 1339, 1344, 1342, 1311, 302, 1531, 1539, 1526, 1545, 1560, 1533]

# Konsentratsiyani hisoblash
konsentratsiya = [s / h for s, h in zip(spirt_miqdor, havo_miqdor)]

# Vaqt bo'yicha konsentratsiyani plot qilish
vaqt = list(range(len(spirt_miqdor)))

plt.plot(vaqt, konsentratsiya, marker='o', linestyle='-')
plt.xlabel('Vaqt (taktlar)')
plt.ylabel('Konsentratsiya')
plt.title('Spirt-havo aralashmasi konsentratsiyasi')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Berilgan ma'lumotlar
vaqt = [365, 363, 366, 360, 361, 362, 358, 368, 359, 350, 346, 347, 346, 350, 354, 336]
spirt_miqdor = [1351, 1344, 1340, 1341, 1345, 1339, 1344, 1342, 1311, 302, 1531, 1539, 1526, 1545, 1560, 1533]

# Konsentratsiyani hisoblash
konsentratsiya = [s / h for s, h in zip(spirt_miqdor, vaqt)]

# Grafikni yaratish
plt.plot(vaqt, konsentratsiya, marker='o', linestyle='-')
plt.xlabel('Vaqt')
plt.ylabel('Konsentratsiya')
plt.title('Spirt-havo aralashmasi konsentratsiyasi')
plt.grid(True)

# Grafikni chiqarish
plt.show()
```

```python
import numpy as np

# Berilgan ma'lumotlar (haqiqiy natijalar)
y_true = np.array([1351, 1344, 1340, 1341, 1345, 1339, 1344, 1342, 1311, 302, 1531, 1539, 1526, 1545, 1560, 1533])
# Bashorat qilingan natijalar
y_pred = np.array([1354.24, 1344.24, 1340.24, 1341.24, 1345.24, 1339.42, 1344.42, 1342.38, 1311.38, 302.38, 1531.38, 1539.38, 1526.38, 1545.38, 1560.38, 1533.38])

# O'rta kvadrat xato (MSE) hisoblash
mse = np.mean((y_true - y_pred) ** 2)

print("O'rta kvadrat xato (MSE):", mse)
```
