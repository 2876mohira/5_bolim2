```python
Xmean = np.mean(X)
ymean = np.mean(y)
theta1 = sum((X-Xmean)*(y-ymean))/sum((X-Xmean)**2)
print(f"{theta1=}")
```

```python
theta0 = ymean - theta1*Xmean
print(f"{theta0=}")
```

```python
x_test = df.sample(10, random_state=50)['Sales'].to_numpy()
print(f"{x_test=}")
y_test = df.sample(10, random_state=50)['TV'].to_numpy()
print(f"{y_test=}")
```

```python
y_predict = theta0 + theta1*x_test
print(f"{y_predict=}")
```
